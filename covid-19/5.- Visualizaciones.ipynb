```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
path = os.getcwd()
path
```

```python
path_tad = path+"/TAD_COVID19.csv"
df_tad = pd.read_csv(path_tad)
```

```python
"""Grafica de barras cdmx"""
eje_x = list(cdmx["Municipio"])
eje_y= list(cdmx["Caso de sars-cov-2  confirmado"])
plt.bar(eje_x, eje_y)
plt.title("Contagios Confirmados Covid 19 Historico CDMX")
plt.xlabel("Municipios")
plt.ylabel("Contagios")
plt.xticks(rotation=85)
plt.grid(True)
plt.savefig('Contagios Confirmados Covid 19 Historico CDMX.jpg',bbox_inches='tight')
plt.show()
```

```python
"""Grafica de barras edomes"""
eje_x = list(edomex["Municipio"])
eje_x = eje_x[:20]
eje_y= list(edomex["Caso de sars-cov-2  confirmado"])
eje_y= eje_y[:20]
plt.bar(eje_x, eje_y, color = "green")
plt.title("Contagios Confirmados Covid 19 Historico EDOMEX (Primeros 20 Municipios)")
plt.xlabel("Municipios")
plt.ylabel("Contagios")
plt.xticks(rotation=85)
plt.grid(True)
plt.savefig('Contagios Confirmados Covid 19 Historico EDOMEX.jpg',bbox_inches='tight')
plt.show()
```

```python
"""Grafica de pastel"""
enfermedad=[df_tad["Si  Diabetes"].sum(),df_tad["Si  Epoc"].sum(),df_tad["Si  Asma"].sum(),\
            df_tad["Si  Inmusupr"].sum(),df_tad["Si  Hipertension"].sum(),\
            df_tad["Si  Otras complicaciones"].sum(), df_tad["Si  Cardiovascular"].sum(),\
           df_tad["Si  Obesidad"].sum(),df_tad["Si  Renal cronica"].sum(),df_tad["Si  Tabaquismo"].sum()]
```

```python
enfermedad
```

```python
enfermedades_name=["Diabetes","Epoc","Asma","Inmunosupresión","Hipertensión","Otras enfermedades",\
                  "Cardiovascular","Obesidad","Renal cronica","Tabaquismo"]
```

```python
plt.pie(enfermedad, labels = enfermedades_name,pctdistance=1)
plt.title("Enfermedades en pacientes registrados")
plt.axis("equal")
plt.savefig('Enfermedades en pacientes registrados.jpg',bbox_inches='tight')
plt.show()
```

```python
plt.bar(enfermedades_name, enfermedad, color = "orange")
plt.title("Enfermedades en pacientes registrados")
plt.xlabel("Enfermedades")
plt.ylabel("Pacientes")
plt.xticks(rotation=85)
plt.grid(True)
plt.savefig('Enfermedades en pacientes registrados_hist.jpg',bbox_inches='tight')
plt.show()
```

```python
hospitalizado = [df_tad["Paciente Hospitalizado"].sum(),df_tad["Paciente Ambulatorio"].sum()]
hospitalizado_name = ["Hospitalizados","Ambulatorios"]
```

```python
"""Contagiados"""

contagiados=[df_tad['Caso de sars-cov-2  confirmado Noviembre 2021'].sum(),\
            df_tad['Caso de sars-cov-2  confirmado Diciembre 2021'].sum(),\
            df_tad['Caso de sars-cov-2  confirmado Enero 2022'].sum(),\
            df_tad['Caso de sars-cov-2  confirmado Febrero 2022'].sum()]
contagiados_name=["Noviembre 2021","Diciembre 2021","Enero 2022","Febrero 2022"]
plt.bar(contagiados_name, contagiados, color = "orange")
plt.title("Contagiados en CDMX y EDOMEX los ultimos 4 meses")
plt.xlabel("Meses")
plt.ylabel("Contagiados")
plt.xticks(rotation=85)
plt.grid(True)
plt.savefig('Contagiados en CDMX y EDOMEX los ultimos 4 meses',bbox_inches='tight')
plt.show()
```

```python
"""Muertos"""

muerto=[df_tad['Muerto Noviembre 2021'].sum(),\
            df_tad['Muerto Diciembre 2021'].sum(),\
            df_tad['Muerto Enero 2022'].sum(),\
            df_tad['Muerto Febrero 2022'].sum()]
muerto_name=["Noviembre 2021","Diciembre 2021","Enero 2022","Febrero 2022"]
plt.bar(muerto_name, muerto, color = "orange")
plt.title("Muertos en CDMX y EDOMEX los ultimos 4 meses")
plt.xlabel("Meses")
plt.ylabel("Muertos")
plt.xticks(rotation=85)
plt.grid(True)
plt.savefig('Muertos en CDMX y EDOMEX los ultimos 4 meses',bbox_inches='tight')
plt.show()
```
